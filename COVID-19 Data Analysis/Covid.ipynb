```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime as dt
```

```python
data=pd.read_csv('datasets_686214_1284378_covid-data.csv')
data.head()
```

```python
data["date"]=pd.to_datetime(data["date"], dayfirst=True).dt.normalize()
data.date
```

```python
#data.set_index("date",inplace=True)
data.info()
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data["total_cases_pm"]=data["total_cases"]/10*6
```

```python
sns.set(style="whitegrid",palette="inferno")
countries=["India","United States","Ireland","Italy","Brazil"]
sns.relplot(data=data,x="date",y="new_cases",kind="line",height=6,hue="location",hue_order=countries)
plt.xlabel("Months")
plt.ylabel("Cases per day")
plt.show()
```

```python
sns.set(style="whitegrid",palette="colorblind")
countries=["India","United States","Ireland","Italy","Brazil"]
sns.relplot(data=data,x="date",y="new_deaths",kind="line",height=6,hue="location",hue_order=countries)
plt.xlabel("Months")
plt.ylabel("Cases per day")
plt.show()
```

```python
sns.set(style="whitegrid",palette="colorblind")
countries=["India","United States","Ireland","Italy","Brazil"]
sns.relplot(data=data,x="date",y="total_cases",kind="line",height=6,hue="location",hue_order=countries)
plt.xlabel("Months")
plt.ylabel("Cases per day")
plt.show()
```

```python
sns.set(style="whitegrid",palette="colorblind")
countries=["India","United States","Ireland","Italy","Brazil"]
sns.relplot(data=data,x="date",y="total_deaths",kind="line",height=6,hue="location",hue_order=countries)
plt.xlabel("Months")
plt.ylabel("Cases per day")
plt.show()
```

```python
sns.set(style="whitegrid",palette="inferno")
countries=["India","United States"]
g=sns.PairGrid(data=data,height=4,hue="location",hue_order=countries,vars=["new_cases","new_deaths"])
g=g.map_diag(plt.hist,alpha=0.4)
g=g.map_offdiag(plt.scatter,alpha=0.5)
g=g.add_legend()
plt.show()
```
