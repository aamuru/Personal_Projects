```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path='faces.jpg'
xml_path='face.xml'
```

```python

img=img=cv2.imread(img_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(img)
plt.show()
```

### Harcascade Classifier

```python
face_cascade=cv2.CascadeClassifier(xml_path)
faces=face_cascade.detectMultiScale(img)
faces.shape
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.figure(figsize=(7,7))
plt.imshow(img)
plt.show()
```

### Eye Detection

```python
xml_path_eye='eye.xml'
img=img=cv2.imread(img_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
face_cascade=cv2.CascadeClassifier(xml_path)
faces=face_cascade.detectMultiScale(img)
eye_cascade=cv2.CascadeClassifier(xml_path_eye)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    single_face=img[y:y+h,x:x+w]
    eyes=eye_cascade.detectMultiScale(single_face,1.5,5)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
plt.imshow(img)
plt.show()
```

```python
##### Detect the smile
smile_xml_path = 'smile.xml'
smile_img_path = 'smiles.jpg'
plt.figure(figsize=(7,7))
img = cv2.imread(smile_img_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
face_cascade = cv2.CascadeClassifier(xml_path)
smile_cascade = cv2.CascadeClassifier(smile_xml_path)
faces = face_cascade.detectMultiScale(img,1.2)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    single_face = img[y:y+h,x:x+w]
    smiles = smile_cascade.detectMultiScale(single_face,1.6,15)
    for (sx,sy,sw,sh) in smiles:
        cv2.rectangle(single_face,(sx,sy),(sx+sw,sy+sh),(0,0,255),2)
plt.imshow(img)
plt.show()
```

## Detect Object by color

```python
import cv2
cam=cv2.VideoCapture(0)
while True:
    ret,img=cam.read()
    hsv=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    cv2.imshow('HSV',hsv)
    cv2.imshow("WebCam",img)
    if cv2.waitKey(1)==13:
        break
cam.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
cam = cv2.VideoCapture(0)  ### open the webcam
while True:
    ret,img = cam.read()
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    ### Red Colour
    low = np.array([140,150,0])
    high = np.array([180,255,255])
    img_mask = cv2.inRange(hsv,low,high);
    output = cv2.bitwise_and(img,img,mask=img_mask)
    cv2.imshow("Output",output)
    cv2.imshow("HSV",hsv)
    cv2.imshow("WebCam",img)
    cv2.imshow("Mask",img_mask)
    if cv2.waitKey(1) == 13:   ## ascii of enter(13)
        break
cam.release()
cv2.destroyAllWindows()
```
